```python
may_desc = may_trip.describe()
june_desc = june_trip.describe()
july_desc = july_trip.describe()
```

```python
may_duration_low_outliers = may_desc.loc['25%', 'tripDuration'] - (1.5 * (may_desc.loc['75%', 'tripDuration']-may_desc.loc['25%', 'tripDuration']))
may_duration_high_outliers = may_desc.loc['75%', 'tripDuration'] + (1.5 * (may_desc.loc['25%', 'tripDuration']-may_desc.loc['25%', 'tripDuration']))
```

```python
may_distance_low_outliers = may_desc.loc['25%', 'tripDistance'] - (1.5 * (may_desc.loc['75%', 'tripDistance']-may_desc.loc['25%', 'tripDistance']))
may_distance_high_outliers = may_desc.loc['75%', 'tripDistance'] + (1.5 * (may_desc.loc['25%', 'tripDistance']-may_desc.loc['25%', 'tripDistance']))
```

```python
june_duration_low_outliers = june_desc.loc['25%', 'tripDuration'] - (1.5 * (june_desc.loc['75%', 'tripDuration']-june_desc.loc['25%', 'tripDuration']))
june_duration_high_outliers = june_desc.loc['75%', 'tripDuration'] + (1.5 * (june_desc.loc['25%', 'tripDuration']-june_desc.loc['25%', 'tripDuration']))
```

```python
june_distance_low_outliers = june_desc.loc['25%', 'tripDistance'] - (1.5 * (june_desc.loc['75%', 'tripDistance']-june_desc.loc['25%', 'tripDistance']))
june_distance_high_outliers = june_desc.loc['75%', 'tripDistance'] + (1.5 * (june_desc.loc['25%', 'tripDistance']-june_desc.loc['25%', 'tripDistance']))
```

```python
july_duration_low_outliers = july_desc.loc['25%', 'tripDuration'] - (1.5 * (july_desc.loc['75%', 'tripDuration']-july_desc.loc['25%', 'tripDuration']))
july_duration_high_outliers = july_desc.loc['75%', 'tripDuration'] + (1.5 * (july_desc.loc['25%', 'tripDuration']-july_desc.loc['25%', 'tripDuration']))
```

```python
july_distance_low_outliers = july_desc.loc['25%', 'tripDistance'] - (1.5 * (july_desc.loc['75%', 'tripDistance']-july_desc.loc['25%', 'tripDistance']))
july_distance_high_outliers = july_desc.loc['75%', 'tripDistance'] + (1.5 * (july_desc.loc['25%', 'tripDistance']-july_desc.loc['25%', 'tripDistance']))
```

```python
may_trip.loc[(may_trip.tripDuration < may_duration_low_outliers) | (may_trip.tripDuration > may_duration_high_outliers)]
```

```python
june_trip.loc[(june_trip.tripDuration < june_duration_low_outliers) | (june_trip.tripDuration > june_duration_high_outliers)]
```

```python
july_trip.loc[(july_trip.tripDuration < july_duration_low_outliers) | (july_trip.tripDuration > july_duration_high_outliers)]
```

```python
may_trip = may_trip.loc[(may_trip.tripDuration > may_duration_low_outliers) & (may_trip.tripDistance > may_distance_low_outliers) & (may_trip.tripDuration < may_duration_high_outliers) & (may_trip.tripDistance < may_distance_high_outliers) ]
```

```python
june_trip = june_trip.loc[(june_trip.tripDuration > june_duration_low_outliers) & (june_trip.tripDistance > june_distance_low_outliers) & (june_trip.tripDuration < june_duration_high_outliers) & (june_trip.tripDistance < june_distance_high_outliers)]
```

```python
july_trip = july_trip.loc[(july_trip.tripDuration > july_duration_low_outliers) & (july_trip.tripDistance > july_distance_low_outliers) & (july_trip.tripDuration < july_duration_high_outliers) & (july_trip.tripDistance < july_distance_high_outliers)]
```
